```python
import json
import re
```

```python
with open("comic_characters_dict.json", "r") as f:
    characters_dict = json.loads(f.read())
```

```python
def remove_parentheses(text):
    return re.sub(r'\s*\([^)]*\)', '', text)

def clean_text(text):
    text = re.sub(r'\[\d+\]', '', text)
    return text
```

```python
stop_words = ["reference", "mention", "appearance", "only in", "on screen", "origin revealed", "flashback", "skeleton", "joins"]
```

```python
for title, characters in characters_dict.items():
    for i, character in enumerate(characters):
        new_name = character["name"]
        for word in stop_words:
            if word in character["name"].lower():
                new_name = remove_parentheses(character["name"])
                old_href = character["href"]

        new_name = clean_text(new_name)
        characters_dict[title][i] = {"name": new_name, "href": old_href}
```

```python
for title, characters in characters_dict.items():
    for i, character in enumerate(characters):
        print(character)
```

```python
with open("cleaned_comic_characters_dict.json", "w") as fp:
    json.dump(characters_dict , fp) 
```
